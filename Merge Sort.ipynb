```python
def mergesort(a):
    #leaf
    n=len(a)
    if n<=1:
        print('Hi a is ',a)
        return
    else:
        print('n is ',n)
        #Internal Node
        mid=n//2
        print('Array is ',a)
        print('Mid index is ',mid)
        suba1=a[:mid]
        suba2=a[mid:]
        print(f'suba1 is {suba1} suba2 is {suba2}')
        mergesort(suba1)
        mergesort(suba2)
    #merge the 2 sorted halves
    
        i=j=k=0
        while i<len(suba1) and j<len(suba2):
            if suba1[i]<suba2[j]:
                a[k]=(suba1[i])
                k+=1
                i+=1
              
            else:
                a[k]=suba2[j]
               
                j+=1
                k+=1

        #gather phase
        while i<len(suba1):
            
                a[k]=(suba1[i])
                k+=1
                i+=1
            
        while j<len(suba2):
                a[k]=(suba2[j])
                k+=1
                j+=1
                
        
        print('sorted a is',a)
    
        


a=[5,4,34,65,356,0,-9,3,1,2]
n=len(a)
mid=n//2
start=0
end=(n-1)
print (f'len is {n} mid= {mid} start= {start} end= {end}')        
mergesort(a)
print('after sort a',a)
```
