```python

def counting_sort(a):
    """
    Args:
     arr(list_int32)
    Returns:
     list_int32
    """
    # Write your code here.
    
        
    #step 1 find max in list
    max=a[0]
    for i in range(0,len(a)):
        if a[i]>max:
            max=a[i]
  
   #find min
    min=a[0]
    for i in range(1,len(a)):
        if a[i]<min:
            min=a[i]


    offset=(0-min)  


#create count array till max+1 and put 0 in all elements
    countarr=[]
    for i in range(offset+min,(offset+max)+1):
        countarr.append(0)

#put frequency of elements of a in countarr
    for i in range(0,len(a)):
        countarr[offset+a[i]]+=1


#build array using countarr
    outarr=[]
    for i in range (0,len(countarr)):
        if countarr[i]!=0:
            for x in range(1,countarr[i]+1):
                outarr.append(i-offset)
#print('sorted output-',outarr)
    
    return outarr
a=[-2,-9,-13,-4,-13,9]
print(f'input is-{a} output is-{counting_sort(a)} ')
```
